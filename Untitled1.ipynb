## Column Details
League - League Name

Season - Season Year YYYY-YYYY

Stage - Stage played

Player - Player Full Name

Team - Team Name

GP - Games Played

MIN - Minutes Played

FGM - Field Goals Made

FGA - Field Goals Attempts

3PM - Three Points Made

3PA - Three Points Attempts

FTM - Free Throws Made

FTA - Free Throws Attempts

TOV - Turnovers( turnover in basketball means losing possession of the ball to the other team, either due to a mistake or their defense)

PF - Personal Fouls

ORB - Offensive Rebounds

DRB - Defensive Rebounds

REB - Rebounds (a rebound is the ball that bounces back after an unsuccessful basket attempt. The player who subsequently catches the ball is awarded the rebound.)

AST - Assists ( passes the ball to a teammate in a way that leads directly to a score by field goal)

STL - Steals (steal is a defensive play where a defender intercepts or takes the ball away from an opponent)

BLK - Blocks (a block shot occurs when a defensive player legally deflects a field goal attempt from an offensive player to prevent a score)

PTS - Points

birth_year - Birth Year

birth_month -Birth Month

birth_date - Birth Date

height - Height [Feet]

height_cm - Height [CM]

weight - Weight [Pounds]

weight_kg - Weight [KG]

nationality - Nationality

high_school - High School

```python
# Define the dropdown widget for player selection
players = data['Player'].unique()
player_dropdown = widgets.Dropdown(options=players, description='Select Player:')

# Global variable to store selected player name
global player_name
player_name = players[0]  # Defaulting to the first player initially

# Observer to update player_name when a player is selected
def on_change(change):
    global player_name
    if change['type'] == 'change' and change['name'] == 'value':
        player_name = change['new']

player_dropdown.observe(on_change)

# Display the dropdown widget
display(player_dropdown)

```

```python
data[data['Player'] == player_name]
```

```python
player_data = data[data['Player'] == player_name]

# (i) Different types of leagues player has played
unique_leagues = player_data['League'].unique()
print("\n(i) Different types of leagues player has played:")
print(unique_leagues)

# (ii) Different Stages
unique_stages = player_data['Stage'].unique()
print("\n(ii) Different Stages player has played:")
print(unique_stages)

# (iii) Different number of teams that player has played
unique_teams = player_data['Team'].unique()
num_unique_teams = len(unique_teams)
print("\n(iii) Number of different teams player has played for:", num_unique_teams)
print("Name of the teams are :",  unique_teams)

# (iv) Total number of games played by the player and number of games played by the player in different teams
total_games_played = player_data['GP'].sum()
print("\n(iv) Total number of games played by the player:", total_games_played)

# Number of games played by the player in different teams
games_played_by_team = player_data.groupby('Team')['GP'].sum()
print("\nNumber of games played by the player in different teams:")
print(games_played_by_team)
```

```python
seasonwise_stats = player_data.groupby('Season').sum()
# Calculate ratios
seasonwise_stats['FGM_FGA_Ratio'] = seasonwise_stats['FGM'] / seasonwise_stats['FGA']
seasonwise_stats['3PM_3PA_Ratio'] = seasonwise_stats['3PM'] / seasonwise_stats['3PA']
seasonwise_stats['FTM_FTA_Ratio'] = seasonwise_stats['FTM'] / seasonwise_stats['FTA']

print("Ratios for each season:")
print(seasonwise_stats[['FGM_FGA_Ratio', '3PM_3PA_Ratio', 'FTM_FTA_Ratio']])
```

```python
plt.figure(figsize=(12, 6))
bar_width = 0.2

# Bar positions
bar_positions = list(range(len(seasonwise_stats.index)))

# Plot FGM/FGA ratio
plt.bar([pos - bar_width for pos in bar_positions], seasonwise_stats['FGM_FGA_Ratio'], bar_width, label='FGM/FGA')

# Plot 3PM/3PA ratio
plt.bar(bar_positions, seasonwise_stats['3PM_3PA_Ratio'], bar_width, label='3PM/3PA')

# Plot FTM/FTA ratio
plt.bar([pos + bar_width for pos in bar_positions], seasonwise_stats['FTM_FTA_Ratio'], bar_width, label='FTM/FTA')

# Set labels and title
plt.xlabel('Season')
plt.ylabel('Ratio')
plt.title('Ratios of Field Goals, Three-Pointers, and Free Throws by Season')
plt.xticks(bar_positions, seasonwise_stats.index, rotation=45)
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
def calculate_metrics(player_name):
    player_data = data[data['Player'] == player_name]
    if player_data.empty:
        print("Player not found.")
        return
    
    # Calculate performance metrics
    fg_percentage = player_data['FGM'].sum() / player_data['FGA'].sum()
    three_point_percentage = player_data['3PM'].sum() / player_data['3PA'].sum()
    free_throw_percentage = player_data['FTM'].sum() / player_data['FTA'].sum()
    points = player_data['PTS'].sum()
    rebounds = player_data['REB'].sum()
    assists = player_data['AST'].sum()
    steals = player_data['STL'].sum()
    blocks = player_data['BLK'].sum()
    turnovers = player_data['TOV'].sum()
    
    # Print or display the calculated metrics
    clear_output(wait=True)  # Clear the previous output
    print("\033[1m Performance Metrics for \033[0m", player_name)
    print("Field Goal Percentage (FG%):", fg_percentage)
    print("Three-Point Percentage (3P%):", three_point_percentage)
    print("Free Throw Percentage (FT%):", free_throw_percentage)
    print("Total Points:", points)
    print("Total Rebounds:", rebounds)
    print("Total Assists:", assists)
    print("Total Steals:", steals)
    print("Total Blocks:", blocks)
    print("Total Turnovers:", turnovers)
    
    
    # Calculate total points scored by the player in different seasons, stages, and leagues
    points_distribution = player_data.groupby(['League', 'Season', 'Stage'])['PTS'].sum().reset_index()
    
    # Print or display the calculated metrics
    print("\033[1m Points scored by", player_name, "in different leagues, seasons, and stages: \033[0m")
    print(points_distribution)
    
    # Create a bar graph
    plt.figure(figsize=(12, 6))
    for league in points_distribution['League'].unique():
        league_data = points_distribution[points_distribution['League'] == league]
        for stage in league_data['Stage'].unique():
            stage_data = league_data[league_data['Stage'] == stage]
            plt.bar(stage_data['Season'] + ' - ' + stage, stage_data['PTS'], label=league + ' - ' + stage)
    
    # Calculate total points scored by the player from field goals, three-pointers, and free throws
    plt.title("Total Points Scored by " + player_name + " in Different Leagues, Seasons, and Stages")
    plt.xlabel("Season - Stage")
    plt.ylabel("Total Points")
    plt.xticks(rotation=45, ha='right')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    fg_points = player_data['FGM'].sum()
    three_points = player_data['3PM'].sum()
    ft_points = player_data['FTM'].sum()

    # Plotting the pie chart using Seaborn
    labels = ['Field Goals', 'Three-Pointers', 'Free Throws']
    sizes = [fg_points, three_points, ft_points]
    plt.figure(figsize=(4, 4))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(labels)))
    plt.title("Point Distribution for {}".format(player_name))
    plt.axis('equal')
    plt.show()
    
    # Plotting the scatter plot using Seaborn
    plt.figure(figsize=(5, 5))
    sns.scatterplot(x='AST', y='PTS', data=player_data, palette='viridis')
    plt.title("Relationship between Points and Assists for {}".format(player_name))
    plt.xlabel("Assists")
    plt.ylabel("Points")
    plt.grid(True)
    plt.show()
```

```python
calculate_metrics(player_name)
```

```python
features = ['AST', 'REB', 'STL', 'BLK']

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot boxplots for each feature
for idx, feature in enumerate(features):
    row = idx // 2
    col = idx % 2
    sns.boxplot(x='Season', y=feature, data=player_data, ax=axes[row, col], palette='viridis')
    axes[row, col].set_title("{} - {}".format(player_name, feature))
    axes[row, col].set_xlabel("Season")
    axes[row, col].set_ylabel(feature)
    axes[row, col].tick_params(axis='x', rotation=45)
    axes[row, col].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
player_data["avg_points_per_game"] = player_data["PTS"]/player_data["GP"]
```

```python
player_team_points = player_data.groupby("Team")["avg_points_per_game"].mean().reset_index()

plt.figure(figsize=(3, 3))
sns.barplot(x='Team', y='avg_points_per_game', data=player_team_points, palette='viridis')
plt.title('Average Points Scored Per Game by player {} ' .format(player_name) )
plt.xlabel('Average Points ')
plt.ylabel('Team')
plt.show()
```
